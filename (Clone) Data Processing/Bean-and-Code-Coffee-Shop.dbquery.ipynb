```sql
select
transaction_id,
transaction_date,
transaction_time,
transaction_qty,
store_id,
store_location,
product_id,
unit_price,
product_category,
product_type,
product_detail,
 product_category,
 
---Now we adding columns to enhance the table using ALIAS (AS FUNCTIONS)

dayname(transaction_date) As day_name,
monthname(transaction_date) As month_name,
dayofmonth(transaction_date) As day_of_month,

---Naming the with name of the days (monday-sunday)

 CASE
 WHEN dayname(transaction_date) IN ('Sat', 'Sun') THEN 'Weekend'
WHEN dayname(transaction_date) IN ('Mon', 'Tue', 'Wed', 'Thu', 'Fri') THEN 'Weekday'
END AS day_classification,

--- Adding time buckets (How to extract time from transaction time column)
 CASE 
    WHEN date_format(transaction_time,'HH:MM:SS') BETWEEN '05:00:00' AND '08:59:59' THEN '01 Rush Hour'
    WHEN date_format(transaction_time,'HH:MM:SS') BETWEEN '09:00:00' AND '11:59:59' THEN '02 Mid Morning'
    WHEN date_format(transaction_time,'HH:MM:SS') BETWEEN '12:00:00' AND '14:59:59' THEN '03 Afternoon'
    WHEN date_format(transaction_time,'HH:MM:SS') BETWEEN '15:00:00' AND '18:59:59' THEN '04 Night'
    ELSE '05.Night'
    END AS time_classification,
   
---Adding spending buckets (can be used for discounts, sales, specials but one get second item 50% off)

 CASE
  WHEN (transaction_qty*unit_price) <= 50 THEN '01.Low Spend'
  WHEN (transaction_qty*unit_price) BETWEEN 51 AND 200 THEN '02.Medium Spend'
  WHEN (transaction_qty*unit_price) BETWEEN 201 AND 300 THEN '03.Moreki'
 ELSE '04.Blesser'
 END AS spend_bucket,

---Adding Revenue
transaction_qty*unit_price As revenue
FROM `workspace`.`default`.`bright_coffee_shop_sales`
```
